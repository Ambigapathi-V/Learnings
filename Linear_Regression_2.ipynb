```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/hiring.csv")
df
```

```python
df.rename(columns={'test_score(out of 10)':'test_score','interview_score(out of 10)':'interview_score','salary($)':'salary'},inplace=True)
```

```python
df.dtypes
```

```python
df.dtypes
```

```python
import math
med_testscore=math.floor(df.test_score.median())
med_testscore
```

```python
df['test_score']=df['test_score'].fillna(med_testscore).astype('int')
df
```

```python
df['experience']=df['experience'].fillna('zero')
df
```

```python
from word2number import w2n
```

```python
df.experience=df.experience.astype(str)
df.experience= df.experience.apply(w2n.word_to_num)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score','interview_score']],df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
